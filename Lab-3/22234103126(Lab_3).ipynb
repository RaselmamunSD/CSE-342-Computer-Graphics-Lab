```python
import matplotlib.pyplot as plt

def midpoint_line(x1, y1, x2, y2):
    points = []
    dx = x2 - x1
    dy = y2 - y1


    if abs(dy) <= abs(dx):
        d = dy - (dx / 2)
        x, y = x1, y1
        points.append((x, y))

        while x < x2:
            x += 1
            if d < 0:
                d += dy
            else:
                d += dy - dx
                y += 1
            points.append((x, y))


    else:
        d = dx - (dy / 2)
        x, y = x1, y1
        points.append((x, y))

        while y < y2:
            y += 1
            if d < 0:
                d += dx
            else:
                d += dx - dy
                x += 1
            points.append((x, y))

    return points


def plot_line(x1, y1, x2, y2):
    points = midpoint_line(x1, y1, x2, y2)


    plt.figure(figsize=(8, 6))
    plt.grid(True)
    plt.axis('equal')


    for (x, y) in points:
        plt.fill_between([x, x+1], y, y+1, color="purple", alpha=0.6)


    plt.plot([x1+0.5, x2+0.5], [y1+0.5, y2+0.5], color="red")

    plt.xlim(0, max(x1, x2) + 5)
    plt.ylim(0, max(y1, y2) + 5)
    plt.show()



plot_line(3, 2, 17, 5)
```
