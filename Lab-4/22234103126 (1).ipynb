```python
import matplotlib.pyplot as plt

# Bresenham's Midpoint Circle Algorithm
def bresenham_circle(xc, yc, r):
    x = 0
    y = r
    d = 3 - 2 * r
    points = []

    while y >= x:
        # 8 symmetric points
        points.extend([
            (xc + x, yc + y),
            (xc - x, yc + y),
            (xc + x, yc - y),
            (xc - x, yc - y),
            (xc + y, yc + x),
            (xc - y, yc + x),
            (xc + y, yc - x),
            (xc - y, yc - x)
        ])

        if d < 0:
            d = d + 4 * x + 6
        else:
            d = d + 4 * (x - y) + 10
            y -= 1
        x += 1

    return points


# 2. Multiple Concentric Circles
def concentric_circles(xc, yc, radii):
    plt.figure(figsize=(6,6))
    for r in radii:
        points = bresenham_circle(xc, yc, r)
        x_vals, y_vals = zip(*points)
        plt.scatter(x_vals, y_vals, s=10, label=f"Radius {r}")
    plt.gca().set_aspect('equal', adjustable='box')
    plt.legend()
    plt.title("Concentric Circles with Bresenham's Algorithm")
    plt.show()



# 3. Shifted Circle

def shifted_circle():
    plt.figure(figsize=(6,6))
    points = bresenham_circle(30, 40, 15)
    x_vals, y_vals = zip(*points)
    plt.scatter(x_vals, y_vals, s=10, color="red")
    plt.gca().set_aspect('equal', adjustable='box')
    plt.title("Shifted Circle at (30,40)")
    plt.show()



# 4. Circle in Each Quadrant

def quadrant_circles():
    plt.figure(figsize=(6,6))
    centers = [(30,30), (-30,30), (-30,-30), (30,-30)]
    for xc, yc in centers:
        points = bresenham_circle(xc, yc, 15)
        x_vals, y_vals = zip(*points)
        plt.scatter(x_vals, y_vals, s=10)
    plt.gca().set_aspect('equal', adjustable='box')
    plt.title("One Circle in Each Quadrant (radius=15)")
    plt.show()



# 5. Filled Circle

def filled_circle(xc, yc, r):
    plt.figure(figsize=(6,6))
    for radius in range(r+1):
        points = bresenham_circle(xc, yc, radius)
        x_vals, y_vals = zip(*points)
        plt.scatter(x_vals, y_vals, s=5, color="blue")
    plt.gca().set_aspect('equal', adjustable='box')
    plt.title("Filled Circle (radius=15)")
    plt.show()



# 6. Olympic Rings

def olympic_rings():
    plt.figure(figsize=(8,6))
    colors = ["blue", "black", "red", "yellow", "green"]
    centers = [(0,0), (40,0), (80,0), (20,-30), (60,-30)]
    radius = 20

    for (xc, yc), color in zip(centers, colors):
        points = bresenham_circle(xc, yc, radius)
        x_vals, y_vals = zip(*points)
        plt.scatter(x_vals, y_vals, s=8, color=color)

    plt.gca().set_aspect('equal', adjustable='box')
    plt.title("Olympic Rings using Bresenham's Algorithm")
    plt.show()



# Run all parts

concentric_circles(0, 0, [5, 10, 15])
shifted_circle()
quadrant_circles()
filled_circle(0, 0, 15)
olympic_rings()
```
